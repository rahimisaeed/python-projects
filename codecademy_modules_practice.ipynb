```python
import random as rand
from matplotlib import pyplot as plt
from decimal import Decimal
from datetime import datetime
```

```python
random_list = []

# generate a list of random numbers between 1 and 100
random_list = [rand.randint(i, i+1) for i in range(101)]
print(random_list)
```

```python
random_number = rand.choice(random_list)
print(random_number)
```

```python
numbers_a = range(1, 13)
numbers_b = rand.sample(range(1000), 12)
plt.plot(numbers_a, numbers_b)
plt.show()
```

```python
no_decimal = 0.2 + 0.69
print(no_decimal)

with_decimal = Decimal('0.2') + Decimal('0.69')
print(with_decimal)
```
